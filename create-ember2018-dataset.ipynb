Intialize new object for dataset

```python
ds = {}
ds["train"] = []
ds["test"] = []
```

Convert train ds to string

```python
count = 0
for elem in train_list:
  count = count + 1
  print(f"Count: {count}")
  temp_obj = {}
  y = elem.pop()
  x = " ".join(str(ele) for ele in elem)
  temp_obj["input"] = x
  temp_obj["label"] = y
  ds["train"].append(temp_obj)
```

Writes the Datasets as many smaller files

```python
!ls drive/MyDrive/ember2018-data/
```

```python
import os
file_path = "drive/MyDrive/ember2018-data/data"
```

```python
def write_ds(ds, file_name, div):
  num_files = len(ds) // div
  for i in range(div):
    start = i * num_files
    end = (i + 1) * num_files
    new_file_name = f"{file_name}_{i+1}.jsonl"
    new_path = os.path.join(file_path, new_file_name)
    with jsonlines.open(new_path, "w") as writer:
      if i != div-1:
        writer.write_all(ds[start:end])
      else:
        writer.write_all(ds[start:])
```

```python
train_div = 400
test_div = 100

train_file_name = "ember2018-train"
test_file_name = "ember2018-test"
```

```python
write_ds(ds["train"], train_file_name, train_div)
```

```python
write_ds(ds["test"], test_file_name, test_div)
```
